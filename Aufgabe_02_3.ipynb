```python
### Aufgabe_02_3.ipynb ###
# Matplotlib-Bibliothek fuer die Darstellung
import matplotlib.pyplot as plt
# random-Bibliothek importieren
import random
import numpy as np

# Startwert
position = 30
walk = [position]

# Schritte
#steps = 100

# Erzeugung der Schrittfolge
#for i in range(steps):
    #step = 1 if random.randint(0, 1) else -1
    #position += step
    #walk.append(position)

steps = 100
draws = np.random.randint(0,2,size=steps)
#print(draws)
steps = np.where(draws > 0, 1,-1)
#print(steps)
walk = steps.cumsum() + position
walk = np.insert(walk,0,position) #bei insert haben wir nachher nicht 101 elementen

choice = np.choice([-1,1], size = steps)
walk = np.append(position,choice) #jetzt 101 elemente
walk = np.cumsum(walk)


# Darstellung mit Matplotlib
plt.title("Random Walk", size=12, fontweight='bold')
plt.xlabel("Schritt", size=12, fontweight='bold')
plt.ylabel("Wert", size=12, fontweight='bold')
plt.plot(walk)
plt.show()
```

Ausgabe des Werts an der ersten und letzten Position,
Ausgabe Minimum und das Maximum der Schrittfolge

```python
print(f'{walk[0]} {walk[-1]} {walk.min()} {walk.max()}')
```

```python
### Aufgabe_02_3.ipynb ###
# Matplotlib-Bibliothek fuer die Darstellung
import matplotlib.pyplot as plt
# random-Bibliothek importieren
import random
import numpy as np

# Startwert
position = 30
walk = [position]

# Schritte
steps = 100

# Erzeugung der Schrittfolge
for i in range(steps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)


# Darstellung mit Matplotlib
plt.title("Random Walk", size=12, fontweight='bold')
plt.xlabel("Schritt", size=12, fontweight='bold')
plt.ylabel("Wert", size=12, fontweight='bold')
plt.plot(walk)
plt.show()
```
